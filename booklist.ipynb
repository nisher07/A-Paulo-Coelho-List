```python
import requests
from bs4 import BeautifulSoup, SoupStrainer
import re
import pandas as pd
```

```python
page = 1

book_title = []
rating = []
rate_count = []
year = []
edition = []

while page != 9:
      url = f"https://www.goodreads.com/author/list/566.Paulo_Coelho?page={page}&per_page=30"
      #print(url)
      response = requests.get(url)
      page_content = response.text
      doc = BeautifulSoup(page_content, 'html.parser')

      #book name OK
      a_tags = doc.find_all('a', {'class': 'bookTitle'})
      for tag in a_tags:
        book_title.append(tag.get_text(strip=True))

      #rating OK
      rate = []
      for i in doc.find_all('span', {'class': 'minirating'}):
        unwanted = i.find('span', {'class': 'stars staticStars notranslate'})
        unwanted.extract()
        rate.append(i.text.strip())
      
      for i in rate:
        rating.append(i[:4])
      
      #rate count OK
      for i in rate:
        rate_count.append(i[18:-7].strip())

      #publish year OK
      tmp = doc.find_all('span', {'class': 'greyText smallText uitext'})

      tm =[]
      for t in tmp:
        #unwanted = t.find('span', {'class': 'stars staticStars notranslate'})
        #unwanted.extract()
        tm.append(t.text.strip())
      
      tm2 = []
      for i in tm:
        ttt = i.replace('\n','')
        tm2.append(ttt.strip())

      for i in tm2:
        if "published" in i:
          some = i[70:94].strip()
          x = re.findall("[0-9]+", some)
          year.append(x)
        else:
          year.append("Nan")
        
      #edition OK
      for t in doc.find_all('span', {'class': 'greyText smallText uitext'}):
        ed = t.find_all('a', {'class': 'greyText'})
        if ed != None:
          edition.append(ed)
        else:
          print("nan")

      
      page = page + 1
```

```python
book_dict = {
    'BookName': book_title,
    'Rating': rating,
    'RateCount': rate_count,
    'PublishYear': year,
    'Editions': edition
}
```

```python
book_df = pd.DataFrame(book_dict)
book_df[:5]
```
